```python
import pandas as pd
```

### product data

```python
df=pd.read_excel("DimTables.xlsx",sheet_name='DimProduct')
df
```

```python
df.isnull().sum()
```

```python
df = df.where(pd.notnull(df), None)
```

```python
df.drop_duplicates()
```

### Connecting database

```python
import mysql.connector as ms
```

```python
db = ms.connect(host="localhost", user="root", password="root", database="retail1",raise_on_warnings=True)
cur= db.cursor()
```

### importing product data into table

```python
to_insert = [tuple(row) for row in df.to_numpy()]

insert_query = 'INSERT INTO dim_product (productkey,productsubcategorykey,product,color,model,subcategory,category)VALUES (%s, %s,%s,%s,%s,%s,%s)'

cur.executemany(insert_query, to_insert)
db.commit()
```
